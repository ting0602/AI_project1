## Spilt dataset

```python
import os
import csv
import random

my_cat_path = './my_cat'
not_my_cat_path = './not_my_cat'

my_cat_file_names = os.listdir(my_cat_path)
not_my_cat_file_names = os.listdir(not_my_cat_path)

random.shuffle(my_cat_file_names)
random.shuffle(not_my_cat_file_names)

# let train:test = 7:3
num_train = int(len(my_cat_file_names) * 0.7)
n_num_train = int(len(not_my_cat_file_names) * 0.7)

# split to train and test
train_files = my_cat_file_names[:num_train]
test_files = my_cat_file_names[num_train:]

n_train_files = not_my_cat_file_names[:n_num_train]
n_test_files = not_my_cat_file_names[n_num_train:]


# set train.csv
with open('train.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)

    # write filename(my_cat) into train.csv -> True
    for file_name in train_files:
        writer.writerow([file_name, '1'])

    # write filename(not_my_cat) into train.csv -> False
    for file_name in n_train_files:
        writer.writerow([file_name, '0'])

# set test.csv
with open('test.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)

    # write filename(my_cat) into test.csv -> True
    for file_name in test_files:
        writer.writerow([file_name, '1'])

    # write filename(not_my_cat) into test.csv -> False
    for file_name in n_test_files:
        writer.writerow([file_name, '0'])
```

```python
import shutil

train_files += n_train_files
test_files += n_test_files
train_path = "./train"
test_path = "./test"

# clear and create new train and test dir
if os.path.exists(train_path):
    shutil.rmtree(train_path)
if os.path.exists(test_path):
    shutil.rmtree(test_path)
os.makedirs(train_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# move images into train and test dir
for root, dirs, files in os.walk(my_cat_path):
    for file in files:
        if file in train_files:
            src_path = os.path.join(root, file)
            dst_path = os.path.join(train_path, file)
            shutil.copy(src_path, dst_path)
        elif file in test_files:
            src_path = os.path.join(root, file)
            dst_path = os.path.join(test_path, file)
            shutil.copy(src_path, dst_path)

for root, dirs, files in os.walk(not_my_cat_path):
    for file in files:
        if file in train_files:
            src_path = os.path.join(root, file)
            dst_path = os.path.join(train_path, file)
            shutil.copy(src_path, dst_path)
        elif file in test_files:
            src_path = os.path.join(root, file)
            dst_path = os.path.join(test_path, file)
            shutil.copy(src_path, dst_path)
```
